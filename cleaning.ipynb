```python
# Create the shooting dataframe
shooting_columns = ['date'] + [col for col in df.columns if col.startswith('all_comps/shooting_')]
shooting_df = df[shooting_columns]
shooting_df.columns = shooting_df.columns.str.replace('all_comps/shooting_', '')

# Create the keeper dataframe
keeper_columns = ['date.1'] + [col for col in df.columns if col.startswith('all_comps/keeper_')]
keeper_df = df[keeper_columns]
keeper_df.columns = keeper_df.columns.str.replace('all_comps/keeper_', '')
keeper_df.rename(columns={'date.1': 'date'}, inplace=True)

# Create the passing dataframe
passing_columns = ['date.2'] + [col for col in df.columns if col.startswith('all_comps/passing_')]
passing_df = df[passing_columns]
passing_df.columns = passing_df.columns.str.replace('all_comps/passing_', '')
passing_df.rename(columns={'date.2': 'date'}, inplace=True)

# Create the possession dataframe
possession_columns = ['date.3'] + [col for col in df.columns if col.startswith('all_comps/possession_')]
possession_df = df[possession_columns]
possession_df.columns = possession_df.columns.str.replace('all_comps/possession_', '')
possession_df.rename(columns={'date.3': 'date'}, inplace=True)

shooting_df = shooting_df[shooting_df['comp']=="Premier League"]
keeper_df = keeper_df[keeper_df['comp']=="Premier League"]
passing_df = passing_df[passing_df['comp']=="Premier League"]
possession_df = possession_df[possession_df['comp']=="Premier League"]
```

```python
# Merge the dataframes using the "date" column as the key
merged_df = shooting_df.merge(keeper_df, on=['date', 'time', 'comp', 'round', 'day', 'venue', 'result', 'gf', 'ga', 'opponent', 'xg', 'xga', 'poss', 'pkatt', 'attendance', 'captain', 'formation', 'referee', 'match report', 'notes', 'season', 'team']) \
    .merge(passing_df, on=['date', 'time', 'comp', 'round', 'day', 'venue', 'result', 'gf', 'ga', 'opponent', 'xg', 'xga', 'poss', 'attendance', 'captain', 'formation', 'referee', 'match report', 'notes', 'season', 'team']) \
    .merge(possession_df, on=['date', 'time', 'comp', 'round', 'day', 'venue', 'result', 'gf', 'ga', 'opponent', 'xg', 'xga', 'poss', 'attendance', 'captain', 'formation', 'referee', 'match report', 'notes', 'season', 'team'])


# If you want to reset the index of the merged dataframe
merged_df.reset_index(drop=True, inplace=True)
```

```python
encoded_formations = pd.get_dummies(merged_df['formation'], prefix='formation')

# Step 2: Feature Engineering (optional)
# For example, you can calculate the number of defenders, midfielders, and forwards in the formation
def count_positions(formation):
    try:
        positions = formation.split('-')
        return {
            'defenders': int(positions[0]),
            'midfielders': int(positions[1]),
            'forwards': int(positions[-1])
        }
    except (ValueError, IndexError):
        # Handle invalid formation strings or missing values here
        return {
            'defenders': 0,
            'midfielders': 0,
            'forwards': 0
        }

merged_df['num_defenders'] = merged_df['formation'].apply(lambda x: count_positions(x)['defenders'])
merged_df['num_midfielders'] = merged_df['formation'].apply(lambda x: count_positions(x)['midfielders'])
merged_df['num_forwards'] = merged_df['formation'].apply(lambda x: count_positions(x)['forwards'])

# Step 3: Concatenate the one-hot encoded formations with the original dataframe
merged_df = pd.concat([merged_df, encoded_formations], axis=1)
```

```python
def rolling_averages(group, cols, new_cols):
    group = group.sort_values("date")
    rolling_stats = group[cols].rolling(3, closed='left').mean()
    for i in range(len(cols)):
        group[new_cols[i]] = rolling_stats.iloc[:, i]
    group = group.dropna(subset=new_cols)
    return group
```
